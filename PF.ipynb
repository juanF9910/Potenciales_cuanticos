```python
%matplotlib notebook
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
from IPython import display

from numpy import *
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['figure.figsize'] = 10, 6 #ancho, alto
font = {'weight' : 'bold',
        'size'   : 15}

matplotlib.rc('font', **font)
################################3
datos = np.loadtxt("programa.txt")
datos1 = np.loadtxt("programa1.txt")
datos2 = np.loadtxt("programa2.txt")
datos3 = np.loadtxt("programa3.txt")
y = np.loadtxt("programa4.txt")
###############################3
a=datos3[0]
b=datos3[1]
dx=datos3[2]
N=datos3[3]
################################
Datos=np.split(datos, N)
Datos1=np.split(datos1, N)
Datos2=np.split(datos2, N)
#######################
x=np.arange(a,b,dx)
############################
%matplotlib notebook
#Gráfica
fig1=plt.figure()
ax=fig1.gca()

#Función de actualización

def actualizar1(i): 
    ax.clear()
    ax.plot(x,Datos[i], c='k',lw=2, label='$|\Psi|^2$')
    ax.plot(x,Datos1[i],c='g', lw=0.6, label='$\Psi_{real}$')
    ax.plot(x,Datos2[i],c='r', lw=0.6, label='$\Psi_{im}$')
    ax.plot(x,y,c='b', lw=1, label='$V(x)$')
    plt.axvline(x =a, color = 'y')
    plt.axvline(x = b, color = 'y')
    plt.xlim(a-1,b+1) #para que no varíen los ejes
    plt.ylim(-3,3)
    plt.xlabel('Posición')
    plt.ylabel('$Probabilidad$')
    plt.title('Función densidad de probabilidad para un pozo infinito')
    plt.grid()
    plt.legend()
#Animación 

ani1=animation.FuncAnimation(fig1,actualizar1,range(len(Datos)), interval=1, repeat=False)
#en donde está range se ponen los datos para meter en actualizar
#interval está en milisegundos, cada k milisegundos crea la gráfica


plt.show()
```
